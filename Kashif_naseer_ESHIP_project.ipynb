```python
import pandas as pd
import json
from pandas import json_normalize

# File paths
sapphire_path = r"C:\Users\Kashif Naseer\OneDrive\Desktop\sapphire.json"
khaadi_path = r"C:\Users\Kashif Naseer\OneDrive\Desktop\khaadi.json"

# Load JSON data
with open(sapphire_path, 'r', encoding='utf-8') as f:
    sapphire_data = json.load(f)
sapphire_df = json_normalize(sapphire_data)

with open(khaadi_path, 'r', encoding='utf-8') as f:
    khaadi_data = json.load(f)
khaadi_df = json_normalize(khaadi_data)

# Extract 'firstComment' from both datasets
sapphire_comments = sapphire_df['firstComment'].dropna()
khaadi_comments = khaadi_df['firstComment'].dropna()

# Combine and clean
all_comments = pd.concat([sapphire_comments, khaadi_comments], ignore_index=True)
all_comments = all_comments.drop_duplicates().reset_index(drop=True)

# Final DataFrame
df = pd.DataFrame({'comment': all_comments})

# Display the first few rows
display(df.head())
```

```python
import re

def clean_comment(text):
    text = re.sub(r'[^\w\s]', '', text)  # Remove punctuation
    text = re.sub(r'\d+', '', text)      # Remove digits
    text = text.lower().strip()          # Lowercase and trim
    return text

df['cleaned_comment'] = df['comment'].apply(clean_comment)
display(df.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram of sentiment scores
plt.figure(figsize=(12, 5))
sns.histplot(df['sentiment'], bins=20)
plt.title('Sentiment Distribution of Fashion Comments')
plt.xlabel('Sentiment Score')
plt.ylabel('Count')
plt.savefig(r"C:\Users\Kashif Naseer\OneDrive\Desktop\21F-9306_eship\sentiment_histogram.png")
plt.show()
```

```python
# Sentiment over time
if 'timestamp' in df.columns:
    df['timestamp'] = pd.to_datetime(df['timestamp'])
    time_trend = df.groupby([df['timestamp'].dt.date, 'brand'])['sentiment'].mean().unstack()
    plt.figure(figsize=(12, 5))
    time_trend.plot()
    plt.title('Sentiment Trend Over Time by Brand')
    plt.xlabel('Date')
    plt.ylabel('Average Sentiment')
    plt.savefig(r"C:\Users\Kashif Naseer\OneDrive\Desktop\21F-9306_eship\sentiment_trend.png")
    plt.show()
else:
    print("No 'timestamp' column found. Skipping time trend plot.")
```

```python
# Box plot of sentiment by brand
plt.figure(figsize=(12, 5))
sns.boxplot(x='brand', y='sentiment', data=df)
plt.title('Sentiment Comparison: Sapphire vs. Khaadi')
plt.xlabel('Brand')
plt.ylabel('Sentiment Score')
plt.savefig(r"C:\Users\Kashif Naseer\OneDrive\Desktop\21F-9306_eship\sentiment_boxplot.png")
plt.show()
```

```python
# Categorize sentiment into positive, negative, neutral
def categorize_sentiment(score):
    if score > 0:
        return 'Positive'
    elif score < 0:
        return 'Negative'
    else:
        return 'Neutral'

# Apply categorization
df['sentiment_category'] = df['sentiment'].apply(categorize_sentiment)

# Calculate proportions
sentiment_counts = df['sentiment_category'].value_counts()

# Pie chart
plt.figure(figsize=(8, 8))
plt.pie(sentiment_counts, labels=sentiment_counts.index, autopct='%1.1f%%', startangle=140, colors=['#66b3ff', '#ff9999', '#99ff99'])
plt.title('Sentiment Distribution of Fashion Comments')
plt.savefig(r"C:\Users\Kashif Naseer\OneDrive\Desktop\21F-9306_eship\sentiment_pie_chart.png")
plt.show()
```

```python
# Violin plot of sentiment by brand
plt.figure(figsize=(12, 5))
sns.violinplot(x='brand', y='sentiment', data=df, inner='quartile')
plt.title('Sentiment Polarity Score Distribution by Brand')
plt.xlabel('Brand')
plt.ylabel('Sentiment Polarity Score')
plt.savefig(r"C:\Users\Kashif Naseer\OneDrive\Desktop\21F-9306_eship\sentiment_violinplot.png")
plt.show()
```
